## Axioms of Probability
In order to present axioms, there is a need to introduce the concept of $\sigma-algebras$.  Let us consider the following probability space properties

### Probability Space properties
Let $\Omega$ be an abstract space, that is with no special structure.  Let $2^\Omega$ denote all subsets of $\Omega$, including the empty set denoted by 0.  With $\mathcal{A}$ being a subset of $2^\Omega$, we consider the following properties:
1. $ 0 \in \mathcal{A}$ and $\Omega \in \mathcal{A}$;
2. If $A\in\mathcal{A}\text{ then }A^c \in \mathcal{A},\text{ where }A^c$ denotes the complement of A;
3. $\mathcal{A}$ is closed under finite unions and finite intersections: that is, if $A_1,\dots,A_n$ are all in $\mathcal{A}$ as well (for this it is enough that $\mathcal{A}$ be stable by the union and the intersection of any two sets);
4.  $\mathcal{A}$ is closed under countable unions and intersections: that is, if $A_1, A_2, A_3, \dots$ is a countable sequence of events in $\mathcal{A}$, then $\bigcup_{i=1}^\infty A_i\text{ and }\bigcap_{i=1}^\infty A_i$ are also in  $\mathcal{A}$.

#### Definition 1 An Algebra
 $\mathcal{A}$ is an algebra if it satisfied (1), (2) and (3) probability space properties above.  It is a $\sigma-algebra$ if it satisfied (1), (2) and (4) probability space properties above.

Note that under (2), (1) can be replieced by either (1):$0\in \mathcal{A}$ or by (2) $\Omega \in \mathcal{A}$. Note also that (1)+(4) implies (3), hence $\sigma-algebras$ are in essence algebras but not particularly the reverse.

#### Definition 2 $\sigma$-Algebra
If $\mathcal{C}\subset 2^\Omega$, the $\sigma-algebra$ generated by $\mathcal{C}$, and written $\sigma(\mathcal{C})$, is the smallest $\sigma-algebra$ containing  $\mathcal{C}$.  (It always exists because $2^\Omega$ is a $\sigma-algebra$, and the intersection of a family of $\sigma-algebras$ is again a $\sigma-algebra$.

#### Theorem 1 The Borel $\sigma$-algebra of R
This is generated by intervals from $(-\infty,a]$, where $a\in\mathbb{Q} (\mathbb{Q}=rationals)$
##### Proof
Let $\mathcal{C}$ denote all open intervals.  Since every open set in $\mathbb{R}$ is the countable union of open intervals, we have $\sigma(\mathcal{C})=$ the Borel $\sigma$-algebra of $\mathbb{R}$

Let $\mathcal{D}$ denote all intervals of the form $(\infty,a]$, where $\alpha \in \mathbb{Q}$. Let $(a,b)\in \mathcal{C}$, and let $(a_n)_{n\ge 1}$ be a sequence of rationals decreasing to $\alpha$ and $(b_n)_{n\ge 1}$ be a sequence of rationals increasing strictly to b.  Then
$$\begin{aligned}(a,b) &= \bigcup_{n=1}^\infty(a_n,b_n] \\ &= \bigcup_{n=1}^\infty((a_n,b_n]^c),\end{aligned}$$

Therefore $\mathcal{C} \subset \sigma(\mathcal{D})$, whence $\sigma(\mathcal{C}) \subset \sigma(\mathcal{D})$. However since each element of $\mathcal{D}$ is a closed set, it is also a Borel set, and therefore $\sigma(\mathcal{D})$ is contained in the Borel sets $\mathcal{B}$.  Thus we have
$$\mathcal{B}=\sigma(\mathcal{C})\subset \sigma(\mathcal{D}) \subset \mathcal{B}$$
and hence $\sigma(\mathcal{D})=\mathcal{B}$.

On the state space $\Omega$ the family of all events will always be a $\sigma$-algebra $\mathcal{A}$: the axioms (1), (2) and (3) correspond to the "logical" operations described previously, while axiom (4) is necessary for mathematical reasons.  

#### Definition 3. A probability measure
This is defined on a $\sigma$-algebra $\mathcal{A}$ of $\Omega$ is $\alpha$ function P:$\mathcal{A}\rightarrow[0,1]$ that satisfies:

1. $P(\Omega)=1$
2. For every countable sequence $(A_n)_{n\ge 1}$ of elements of $\mathcal{A}$, pairwise disjoint (that is, $A_n\cap A_m=0$ whenever $n\ne m$), one has
$$P(\bigcup_{n=1}^\infty A_n)=\sum_{n=1}^\infty P(A_n)$$

Axiom (2) above is called **countable additivity**; the number P(A) is called the probability of the event A. Note that:
$$A,B\in\mathcal{A}, A\cap B=0 \implies P(A\cup B)=P(A)+P(B) - - - -\dots(1)$$

This property is called additivity (or finite additivity) and, by an elementary induction, it implies that for every finite $A_1,\dots,A_m$ of pairwise disjoint events $A_i\in\mathcal{A}$ we have
$$P(\bigcup_{n=1}^mA_n)=\sum_{n=1}^mP(A_n)$$
